```python
with open('file.txt') as f:
    lines = [line.strip() for line in f.readlines()]
lines[:5]
```

```python
parsed = [line.split() for line in lines]
parsed
```

```python
scores = []
for hand, bid in parsed:
    cards = {}
    for card in hand:
        if card == "J":
            continue
        if card in cards:
            cards[card] += 1
        else:
            cards[card] = 1
    
    matches = sorted(cards.values(), reverse=True)
    if hand != "JJJJJ":
        matches[0] += hand.count("J")

    print(hand, matches)

        
    if hand == "JJJJJ":
        score = 7
    elif matches[0] == 5:
        score = 7
    elif matches[0] == 4:
        score = 6
    elif matches[0] == 3 and matches[1] == 2:
        score = 5
    elif matches[0] == 3:
        score = 4
    elif matches[0] == 2 and matches[1] == 2:
        score = 3
    elif matches[0] == 2:
        score = 2
    else:
        score = 1

    scores.append((score, hand, bid))
        
    print(cards)

print(scores)
```

```python
map = {
    "A": 14,
    "K": 13,
    "Q": 12,
    "J": 1,
    "T": 10,
}
```

```python
for i in range(len(scores)):
    already_sorted = True
    for j in range(len(scores) - i - 1):
        if scores[j][0] < scores[j+1][0]:
            scores[j], scores[j+1] = scores[j+1], scores[j]
            already_sorted = False
        elif scores[j][0] == scores[j+1][0]:
            for k in range(5):
                val1 = int(scores[j][1][k]) if scores[j][1][k] not in map else map[scores[j][1][k]]
                val2 = int(scores[j+1][1][k]) if scores[j+1][1][k] not in map else map[scores[j+1][1][k]]
                if val1 < val2:
                    scores[j], scores[j+1] = scores[j+1], scores[j]
                    already_sorted = False
                    break
                elif val1 > val2:
                    break

    if already_sorted:
        break

print(scores)

for i, (score, hand, bid) in enumerate(scores):
    if hand == 'J' * 5:
        print(i + 1, bid)
        break
```

```python
sum = 0
for i, (_, _, bid) in enumerate(scores):
    print(len(scores) - i, bid)
    sum += int(bid) * (len(scores) - i)
sum
```
