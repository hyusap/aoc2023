```python
with open('file.txt') as f:
    lines = [line for line in f.readlines()]
lines[:5]
```

```python
uniques = set()
for line in lines:
    for char in line:
        if not char.isdigit() and char != '.':
            # print(char)
            uniques.add(char)
uniques
```

```python
valid_nums = []
gear_matches = {}

for i, line in enumerate(lines):
    print(line)
    cur_number = ''
    gear_loc = None
    for j, char in enumerate(line):
        if char.isdigit():
            cur_number += char
            for oi in range(-1, 2):
                for oj in range(-1, 2):
                    if oi == 0 and oj == 0:
                        continue
                    try:
                        scan = lines[i + oi][j + oj]
                        # print(scan)
                    except IndexError:
                        continue
                    if scan == '*':
                        gear_loc = (i + oi, j + oj)
        else:
            if cur_number != '':
                if gear_loc is not None:
                    if gear_matches.get(gear_loc) is None:
                        gear_matches[gear_loc] = int(cur_number)
                    else:
                        valid_nums.append(gear_matches[gear_loc] * int(cur_number))
                    print('valid', cur_number)
                else:
                    print('invalid', cur_number)
                cur_number = ''
                gear_loc = None
                # print()
                # print(valid_nums)
                

sum(valid_nums)
```
