```python
with open('file.txt') as f:
    lines = [line.strip() for line in f.readlines()]
lines
```

```python
text_nums = ["one", "two", "three", "four", "five", "six", "seven", "eight", "nine"]
```

```python
sum = 0

for line in lines:
    print(line)
    found_nums = []
    
    for i, tnum in enumerate(text_nums):
        if tnum in line:
            print("found", tnum, i+1)
            # found_nums.append([line.find(tnum), i+1])
            last_find = -1
            while True:
                last_find = line.find(tnum, last_find+1)
                if last_find == -1:
                    break
                found_nums.append([last_find, i+1])
    print("found_nums", found_nums)

        
    
    first = True
    nums = []
    for i, c in enumerate(line):
        for index, num in found_nums:
            if i == index:
                nums.append(num)
                break

        if c.isdigit():
            nums.append(int(c))
    print(line, nums)
    val = nums[0] * 10 + nums[-1]
    print("val", val)
    print()
    print()
    sum += val
sum
```
