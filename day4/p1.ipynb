```python
with open('file.txt') as f:
    lines = [line.strip() for line in f.readlines()]
lines[:5]
```

```python
sum_val = 0
for line in lines:
    answers, nums = line.split(": ")[1].split(" | ")
    answers = list(filter(lambda x : x != "", answers.split(" ")))
    nums = list(filter(lambda x : x != "", nums.split(" ")))
    current_value = None
    print(answers, nums)
    for num in nums:
        print(num)

        if num in answers:
            if current_value == None:
                current_value = 1
            else:
                current_value *= 2
    if current_value != None:
        sum_val += current_value
sum_val
```
