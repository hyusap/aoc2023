```python
from tqdm.notebook import tqdm
```

```python
with open('file.txt') as f:
    lines = [line.strip() for line in f.readlines()]
lines[:5]
```

```python
cards = []
for line in lines:
    name, card = line.split(':')
    name = int(name.split()[1])
    answers, nums = card.split(' | ') 
    answers = list(filter(lambda x : x != "", answers.split(" ")))
    nums = list(filter(lambda x : x != "", nums.split(" ")))
    cards.append((answers, nums))
cards
```

```python
total_cards = len(cards)

cache = {}

def evaluate_card(card_index):
    if card_index in cache:
        return cache[card_index]
    answers, nums = cards[card_index]
    num_correct = 0
    for num in nums:
        if num in answers:
            num_correct += 1
    final_value = num_correct
    for sub_card_index in range(card_index + 1, card_index + num_correct + 1):
        final_value += evaluate_card(sub_card_index)

    cache[card_index] = final_value
    return final_value

for i in tqdm(range(len(cards))):
    total_cards += evaluate_card(i)
total_cards
```
